```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
```

```python
df=pd.read_csv("dataset/student_scores.csv")
df.head(3)
```

```python
y=df["Scores"]
x=df[["Hours"]] 
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.show()
```

```python
lr=LinearRegression()
model=lr.fit(x,y)
model.score(x,y)
```

```python

alphas=[1,10,50,100,300,700,1000]
#plt.style.use("fivethirtyeight")
#plt.figure(figsize=(8, 6))
#plt.scatter(x, y, color="blue", label="Gerçek Değerler")  # Gerçek değerler
for i in alphas:
    r = Ridge(alpha=i)
    modelr = r.fit(x, y)
    skor = modelr.score(x, y)
    print(f"Alpha: {i}")
    print("Skor:", skor)
    print("Katsayılar:", modelr.coef_)
    
    # Tahmin çizgisi
    #plt.plot(x, modelr.predict(x), linewidth=2, label=f"Alpha={i}")

# Grafik ayarları
#plt.title("Ridge Regression Tahminleri (Farklı Alpha Değerleri)")
#plt.xlabel("X")
#plt.ylabel("Y")
#plt.legend()
#plt.show()
```
