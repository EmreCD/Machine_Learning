```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#data frame
df=pd.read_csv("dataset/audi.csv")
```

```python
df.head(5) #data frame çalışıyo mu bak bakalım
```

PREPROCESSING

```python
df=df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score'])
df
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat","ppy"]
df
```

```python
df['motor']=df['motor'].str.replace("L","")
df
```

```python
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True)
df['motor']=pd.to_numeric(df['motor'])
df["yil"]=df["yil"].astype(int)
df["mil"]=df["mil"].astype(int)
df["fiyat"]=df["fiyat"].astype(int)
df["vites_Manual"]=df["vites_Manual"].astype(int)
df["yakit_Petrol"]=df["yakit_Petrol"].astype(int)

df
```

```python
y=df[['fiyat']]
x=df.drop("fiyat",axis=1)
lm=LinearRegression()
model=lm.fit(x,y)
df
```

```python
model.predict([[2017,30000,1.6,110.2,2,2600.2,0,1]]) 
#çift [[]] !!!!  :(
```

```python
model.score(x,y)
#doğruluk oranı
```
